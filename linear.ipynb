```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample data: House size (square feet) and corresponding price (in $1000s)
X = np.array([600, 650, 700, 750, 800, 850, 900, 950, 1000]).reshape(-1, 1)  # Feature (House size)
y = np.array([150, 160, 170, 180, 190, 200, 210, 220, 230])  # Target (Price)

print(X,y)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict for a new house size
new_size = np.array([[810]])
predicted_price = model.predict(new_size)

print(predicted_price)
```

```python
print(f"Predicted price for a 810 sq.ft house: Rs {predicted_price[0]}k")
```

```python
# Plot the data and regression line
plt.scatter(X, y, color='blue', label="Actual Prices")
plt.plot(X, model.predict(X), color='red', label="Regression Line")
plt.scatter(new_size, predicted_price, color='green', label="Predicted Price", marker='*', s= 100)
plt.xlabel("House Size (sq.ft)")
plt.ylabel("Price ($1000s)")
plt.legend()
plt.title("Linear Regression - House Price Prediction")
plt.show()
```
